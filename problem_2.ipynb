# Student Exam Scores

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mode
```

```python
array  = [78, 85, 92, 88, 76, 95, 89, 84, 73, 91,
          87, 82, 94, 79, 81, 90, 90, 90, 83, 80,
          75, 97, 98, 72, 74, 96, 93, 99, 95, 100]
scores = np.array(array)
# mean
mean = np.mean(scores)
print("mean of scores : ",mean)
# Median
median = np.median(scores)
print("median of scores : ",median)
# mode
print("mode of scores : ",mode(scores))
# min value
min_value = np.min(scores)
print("minimum scores : ",min_value)
# max_value
max_value = np.max(scores)
print("max scores : ",max_value)
# range 
range = max_value - min_value
print("range : ",range)
# varience
varience = np.var(scores)
print("population varience : ", varience.round(2))
# standard deviation
std_varience = np.std(scores)
print("standard deviation of varience : ", std_varience)
```

# Quartiles

```python
# Q1
Q1 = np.quantile(scores,0.25)
print("quartile one : ", Q1)
# Q2 or median
Q2 = np.quantile(scores,0.50)
print("quantile two : ",Q2)
# Q3 
Q3 = np.quantile(scores,0.75)
print("quantile three : ", Q3)

# IQR
IQR = Q3 - Q1
print("inter quartile range : ",IQR)
```

```python
# calculating z score of the student who scord 97 
mean = np.mean(scores)
std = np.std(scores)
x=97
z_score = (x -mean)/std
print("z_score for 97 : ",z_score.round(2))
```

```python
# histogram
plt.figure(figsize=(8,8))
plt.hist(scores,bins=7,color="purple",edgecolor="black")
plt.xlabel("scores")
plt.ylabel("frequency")
plt.title("Scores Distribution")
plt.show()
```

```python
 # boxplot
plt.figure(figsize=(8,7))
plt.boxplot(scores,vert=False)

plt.title("Box Plot of Student Scores")
plt.xlabel("Scores")

plt.show()
```

```python
# Probability of score greater than 90
total = len(scores)
greater_90 = np.sum(scores >  90)
p_greater_90 = greater_90 / total
print("probablity of greater than 90 : ",p_greater_90.round(3))


# probablity of scores  is less than 80
less_80 = np.sum(scores <= 80)
p_less_80 = less_80 / total
print("probablity is less than 80 ",p_less_80.round(4))

# probablity of scores between 80 and 90 
bet_80_and_90 = np.sum((scores >  80)  & (scores  < 90))
p_bet_80_and_90 = bet_80_and_90/total
print("probablity between 80 and 90 : ",p_bet_80_and_90)
```
