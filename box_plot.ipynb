```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/home/jawad/Downloads/tip.csv")
df
```

# # box plot 

```python
plt.figure(figsize=(8,6))
plt.boxplot(df["total_bill"],vert=False,showfliers=True)
plt.title("box plot")
plt.xlabel("total bill")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["total_bill"],edgecolor="black",color="skyblue")
plt.title("distribution of total bill")
plt.xlabel("total bill")
plt.ylabel("frequency")
plt.show()
```
