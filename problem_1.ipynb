```python
data = [10, 18, 15, 12, 20, 30, 25, 22, 28]
data_sort = np.sort(data)

q1 = np.quantile(data_sort,(0.25))
q2 = np.quantile(data_sort,(0.50))
q3 = np.quantile(data_sort,(0.75))
print("quartile ist : ",q1)
print("quartile 2nd : ",q2)
print("quartile 3rd",q3)
IQR = q3 - q1
print("inter quartile range  :",IQR)

```

```python
dataset = [10, 18, 15, 12, 20, 30, 25, 22, 28]
dataset_sort = np.sort(dataset)
q1 = np.quantile(dataset_sort,(0.25))
q3 = np.quantile(dataset_sort,(0.75))
IQR = q3 - q1
print("inter quartile range  :",IQR)
lower_half = q1 - (1.5 * IQR)
upper_half = q3 + (1.5 * IQR)
outliers=[]
for i in dataset:
    if ( i < lower_half ) | (i > upper_half):
      outliers.append(i)
      
print(outliers)



```
