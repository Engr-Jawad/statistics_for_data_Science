# detect outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

## Define our dataset
dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
outliers = []
def detect_outliers(dataset):
    threshold = 3 # third standard deviation
    mean = np.mean(dataset)
    std = np.std(dataset)
    for i in dataset:
        # z_score formula 
        z_sore = (i-mean)/std
        if np.abs(z_sore>threshold):
            outliers.append(i)
    return outliers
    
print(detect_outliers(dataset))
```

# IQR inter quartile range 
1.sort the data<br>
2.find Q1 and Q3<br>
3.IQR(Q3-q1)<br>
4.lower bound [Q1-1.5(IQR)]<br>
5.upper bound [Q3 + 1.5(IQR)]



```python
sort = np.sort(dataset)
Q1 = np.quantile(dataset,0.25)
Q3 = np.quantile(dataset,0.75)
IQR = Q3 - Q1
outliers= []
print("inter quartile range : ",IQR)
# lowerbound
lowerbound = Q1 - (1.5 * IQR)
# upperbound
upperbound = Q3 + (1.5 * IQR)
for i in dataset:
    if (i < lowerbound ) | (i > upperbound):
       outliers.append(i)
   
print(outliers)
```
